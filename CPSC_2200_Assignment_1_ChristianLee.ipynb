##What is binary classification?
Binary classification is a form of supervised machine learning and statistical classification, and its main objective is to assign each input example to one of two classes. Inputs(x) can vary in terms of form: numbers, text, images вЂ“ but ultimately, will output a вЂњlabelвЂќ(y) that is either always вЂњyes/noвЂќ, вЂњ0/1вЂќ or вЂњ-1/+1вЂќ. Models usually represent this concept of binary classification by finding the probability of the positive class, or in other words, the probability that the output is вЂњyesвЂќ or вЂњ1вЂќ given x, or the input. Often, if that probability is greater than 0.5, you predict class 1 вЂњyesвЂќ(if less, you predict class 0 вЂњnoвЂќ. Commonly used algorithms include Logistic Regression, Support Vector Machine (SVM), Random Forest & Decision Trees, Neural Networks, etc.

##Why does binary classification typically use:

A sigmoid activation in the output layer?
Binary crossentropy as the loss function?
Sigmoid, as the activation function, enables the modelвЂ™s final score to be interpreted as a probability-like number for these binary outcomes. It essentially maps any real number to (0, 1) and we interpret this probability as the probability of y=1 given x, abiding by our decision rule of, for instance, a 0.5 threshold. Binary crossentropy works well with binary classification because itвЂ™s the loss you get when you assume Bernoulli labels (0/1); it heavily penalizes confident, yet wrong predictions and provides strong gradients for learning.

##Why can accuracy be misleading for imbalanced datasets?

Accuracy can be misleading for imbalance datasets because it can be dominated by a majority class. In a dataset of 1000 values where 990 are negative and 10 are positive and the model predicts negative on all of them, accuracy will be 99%. ## Part 1. Model Variations for Binary Classification (IMDb Dataset)

### Loading IMDB Dataset

````python
import keras
from keras.datasets import imdb

# Load the IMDB dataset, limiting it to the top 10,000 most frequent words
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
````

### Data Preprocessing

````python
#Prints the shape of train_data
print(train_data.shape)

#Prints the length of train_data
print(len(train_data))

#Prints train labels from 0 to 20, the shape of the labels, and the type of the
#labels
print(train_labels[:20], train_labels.shape, train_labels.dtype)

#Prints the shape of test_data
print(test_data.shape)

#Prints the length of train_data
print(len(test_data))

#Prints test labels from 0 to 20, the shape of the labels, and the type of the
#labels
print(test_labels[:20], test_labels.shape, test_labels.dtype)

#Gets the column values for column '0' in train data
train_data[0]

#Gets the column values for column '0' for train labels
train_labels[0]

#Calculates the maximum word index found in the dataset/checks parameter of
#num_words=10000
max([max(sequence) for sequence in train_data])
````

````python
#Maps words to an integer index through dictionary 'word_index'
word_index = imdb.get_word_index()

#Maps integer index to words through dictionary 'reverse_word_index'
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

# Decodes the review of the first training data.
decoded_review = " ".join(
    [reverse_word_index.get(i - 3, "?") for i in train_data[0]]
)
#Outputs first 100 characters as a check
decoded_review[:100]
````

````python
import numpy as np
#Defines function 'multi_hot_encode' with sequences and num_classes as inputs
def multi_hot_encode(sequences, num_classes):
  #Sets results to an array with the shape given by the length of the sequences
  #and the number of classes, set to zero
  results = np.zeros((len(sequences), num_classes))
  #Sets specific indices of results[i] to 1
  for i, sequence in enumerate(sequences):
    results[i][sequence] = 1.0
  return results

#Vectorizes 'train_data' and 'test_data' with 10,000 classes
x_train = multi_hot_encode(train_data, 10000)
x_test = multi_hot_encode(test_data, 10000)

#Vectorizes 'train_labels' and 'test_labels' from int type to float
y_train = train_labels.astype("float32")
y_test = test_labels.astype("float32")

#Take the first 10,000 samples from original 'x_train' and 'y_train' and creates
#validation set
x_val = x_train[:10000]
y_val = y_train[:10000]
#Use the remaining values in the dataset for the training set
partial_x_train = x_train[10000:]
partial_y_train = y_train[10000:]

#Prints shape of validation data/labels and partial training data/labels
print(f"Validation data shape (x_val): {x_val.shape}")
print(f"Validation labels shape (y_val): {y_val.shape}")
print(f"Partial training data shape (partial_x_train): {partial_x_train.shape}")
print(f"Partial training labels shape (partial_y_train): {partial_y_train.shape}")
````

### Functions to Generate-Compile-Train Model

````python
from keras import layers

#Creates model from 'keras' library with three layers: two with 16 neurons and
#activation of 'relu' and one with 1 neuron and activation sigmoid
model = keras.Sequential(
    [
        layers.Dense(16, activation = "relu"),
        layers.Dense(16, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ]
)
````

````python
#Compiles model with optimizer "rmsprop", loss "binary_crossentropy", and metric
#"accuracy"
model.compile(
    optimizer="rmsprop",
    loss="binary_crossentropy",
    metrics=["accuracy"],
)
````

````python
#Fit model with x_train and y_train, using 20 epochs, a 64 batch size, and 0.2
#validation split
history = model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=256,
    validation_split=0.2
)
````

````python
#Evaluates baseline model on test set
test_loss_baseline, test_acc_baseline = model.evaluate(x_test, y_test)
print(f"\nBaseline Model (2 layers, 16 units, ReLU, binary_crossentropy):")
print(f"  Test Loss: {test_loss_baseline:.4f}")
print(f"  Test Accuracy: {test_acc_baseline:.4f}")

#Stores baseline for comparison
baseline_results = {
    'test_loss': test_loss_baseline,
    'test_acc': test_acc_baseline
}
````

### Comparison Plots

````python
import matplotlib.pyplot as plt

#Creates a dictionary that contains recorded metrics from the model training
history_dict = history.history

#Calls the keys() method on the newly created dictionary, outputting the names
#of the tracked metrics
history_dict.keys()

#Training loss history
loss_values = history_dict["loss"]

#Validation loss history
val_loss_values = history_dict["val_loss"]

#Creates a sequence of numbers from 1 to the total number of epochs
epochs = range(1, len(loss_values) + 1)

#Plot training loss with red dashed line against the epoch numbers
plt.plot(epochs, loss_values, "r--", label="Training Loss")

#Plot validation loss with blue line against the epoch numbers
plt.plot(epochs, val_loss_values, "b", label="Validation Loss")

#Sets title, x-labels, x-axis ticks, y-labels, legend, etc
plt.title("[IMDB] Training and Validation Loss")
plt.xlabel("Epochs")
plt.xticks(epochs)
plt.ylabel("Loss")
plt.legend()
plt.show()


#Training accuracy history
acc = history_dict["accuracy"]

#Validation accuracy history
val_acc = history_dict["val_accuracy"]

#Plot training accuracy with red-dotted line against epoch numbers
plt.plot(epochs, acc, "r--", label="Training Accuracy")

#Plot validation accuracy with blue line against epoch numbers
plt.plot(epochs, val_acc, "b", label="Validation Accuracy")

#Sets title, x-labels, x-axis ticks, y-labels, legend, etc
plt.title("[IMDB] Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.xticks(epochs)
plt.ylabel("Accuracy")
plt.legend()
plt.show()
````

## Part 2.1: Short response (1-2 paragraphs)

Q1) What is a regression problem?

Regression is a supervised learning technique for predicting continuous, numerical output values (prices, temperature, sales) based on a relationship between inputs and a target variable. Examples include forecasting the price of a house using variables like number of bedrooms, location, and square footage, estimating future sales revenue based on historical data, advertising, and seasonal trends, or predicting the exact temperature for a specific day based on humidity, advertising spend, and seasonal trends.


Q2) How is regression different from classification?

Classification predicts categorical labels and is binary in nature, so it's output would be "Y/N" or "Positive/Negative" or "Will Default/Won't Default". Regression is different in that it doesn't strictly abide by a binary-type output and isn't restricted to just two outputs. It's numerical and predicts continous, numerical values and is subject to the more lax constraint of producing an output as long as it's a real number.


Q3) Why is Mean Squared Error (MSE) commonly used for regression?

Mean Squared Error(MSE) measures the average squared difference between actual and predicted values (essentially an averaged standard deviation). It is commonly used for regression because it is mathematically differentiable (able to take the derivative) which applies well to the continuous nature of regression problems. This ultimately allows for effective and efficient optimization through gradient descent. It penalizes error appropriately(large errors are punished heavily) and it mitigates massive, single-point failures in models.


Q4) What does Mean Absolute Error (MAE) measure, and why is it easier to interpret than MSE?

Mean Absolute Error (MAE) measures the average magnitude of errors in a set of predictions; this calculation is done through finding the average of the absolute differences between predied and actual values. It's easier to interpret because it's natively expressed in the same units as the original data as opposed to being squared units. While being less sophisticated, it's more intuitive because you're simply finding the difference between the predicted values and the actual values instead of squaring it. Finally, MAE doesn't disproportionally deal with errors/outliers, it treats all errors proportionally because it doesn't have a square mechanism that amplifies errors the larger they are.


Q5) Why is accuracy not a good metric for regression?

Accuracy is a broad, relatively arbitrary measure for how many of your predictions are right as a simple fraction of right and wrong. It doesn't contain the necessary nuance of measuring how close a prediction is, which is why it's better for binary answer questions compared to continuous, numerical questions.


Q6) Explain the purpose of normalizing input features in regression models.

Some features will have very different scales; for housing, the historical price of a house will have a very different range compared to sqft property size or the number of bathrooms/beds in a house. Normalizing input features in regression models ensure fair contribution of features and ensures no unjustified weighting to larger-scale variables simply because their values are numerically bigger. As a result, training is faster and more stable and regularization works properly.


## Part 2.2: Regression Code

<u>Introduction</u>: In this part of the assignment, we will use the Boston Housing dataset. It is a famous collection of data derived from the U.S. Census Service concerning housing in the area of Boston, Massachusetts, collected in the 1970s. It contains 506 data points with 13 diverse features representing different aspects of the suburbs, such as the per capita crime rate, the average number of rooms per dwelling, the pupil-teacher ratio, and property tax rates. The goal of this regression task is to predict the median value of owner-occupied homes (in thousands of dollars) based on these 13 distinct community attributes.

### Load the Boston Housing dataset from keras.dataset. Split the data into training and test sets.

````python
import tensorflow as tf
from tensorflow.keras import models, layers
import numpy as np
import matplotlib.pyplot as plt

from keras.datasets import boston_housing

#Loads the Boston Housing dataset from keras.datasets
(train_data, train_labels), (test_data, test_labels) = boston_housing.load_data()

#Prints shape of train_data
print(train_data.shape)

#Prints shape of test_data
print(test_data.shape)
````

### Normalize the features using the training data mean and standard deviation.


````python
# It is best practice to calculate parameters only on training data to prevent data leakage
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std

# Apply the same mean and std to the test data
test_data -= mean
test_data /= std
````

### Build a neural network with:
*   Input layer matching feature size
*   Two hidden layers with ReLU activation
*   One output neuron (no activation)

### Compile the model using:
* Optimizer: Rmsprop
* Loss: mse
* Metric: mae

````python
#Creates model using keras.Sequential()
#layers.Input matches feature size by setting shape equal to train_data
#layers.Dense makes two hidden layers with 64 neurons and activation "relu"
#last layer is the output layer with one neuron and no activation
def get_model():
  model = keras.Sequential(
      [
          layers.Input(shape=(train_data.shape[1],)),
          layers.Dense(64, activation="relu"),
          layers.Dense(64, activation="relu"),
          layers.Dense(1)
      ]
  )
  #Compiles model with optimizer "rmsprop", loss "mse", and metrics "mae"
  model.compile(
      optimizer="rmsprop",
      loss="mse",
      metrics=["mae"]
  )
  return model

````

### Train the model for 100 epochs.

````python
model = get_model()
model.fit(
    train_data,
    train_labels,
    epochs=100,
    batch_size=16,
    validation_split=0.2
)
````

### Evaluate the model on the test set.
